Câu 6: Ý nghĩa các toán tử trong Python

- `/` : Chia lấy kết quả là số thực (float).  
  Ví dụ: `5 / 2` kết quả là `2.5`

- `//` : Chia lấy phần nguyên (floor division).  
  Ví dụ: `5 // 2` kết quả là `2`

- `%` : Chia lấy phần dư (modulo).  
  Ví dụ: `5 % 2` kết quả là `1`

- `**` : Lũy thừa (power).  
  Ví dụ: `2 ** 3` kết quả là `8`

- `and` : Toán tử logic "và", trả về True nếu cả hai vế đều đúng.  
  Ví dụ: `(True and False)` kết quả là `False`

- `or` : Toán tử logic "hoặc", trả về True nếu ít nhất một vế đúng.  
  Ví dụ: `(True or False)` kết quả là `True`

- `is` : So sánh hai đối tượng có cùng tham chiếu trong bộ nhớ hay không.  
  Ví dụ:  
  ```python
  a = [1, 2]
  b = a
  c = [1, 2]
  (a is b) # True
  (a is c) #

Câu 8: Các loại lỗi khi lập trình và cách bắt lỗi trong Python

1. Các loại lỗi thường gặp:
- Lỗi cú pháp (SyntaxError):
  Xảy ra khi viết sai quy tắc cú pháp của Python.  
  Ví dụ:  
  ```python
  print("Hello"
Lỗi khi chạy (RuntimeError):
Xảy ra khi chương trình đang chạy, ví dụ chia cho 0, truy cập biến chưa khai báo,...
    a = 5 / 0  # ZeroDivisionError
Lỗi logic:
Chương trình chạy không báo lỗi nhưng kết quả sai do sai thuật toán
2. Cách bắt lỗi trong Python (try...except):

Sử dụng khối try...except để bắt và xử lý lỗi khi chạy chương trình.
Cú pháp:
try:
    # Code có thể gây lỗi
    x = int(input("Nhập số: "))
    y = 10 / x
    print(y)
except ZeroDivisionError:
    print("Không được chia cho 0!")
except ValueError:
    print("Bạn phải nhập số!")
except Exception as e:
    print("Lỗi khác:", e)
else:
    print("Không có lỗi xảy ra.")
finally:
    print("Kết thúc chương trình.")



Câu 9: Giải thích kết quả tính toán của các biểu thức  
Cho các biến với giá trị:  
i1 = 2  
i2 = 5  
i3 = -3  
d1 = 2.0  
d2 = 5.0  
d3 = -0.5  

Kết quả và giải thích:

(a) `i1 + (i2 * i3)`  
= 2 + (5 * -3) = 2 + (-15) = **-13**  
→ Nhân trước, cộng sau.

(b) `i1 * (i2 + i3)`  
= 2 * (5 + -3) = 2 * 2 = **4**  
→ Cộng trong ngoặc trước, nhân sau.

(c) `i1 / (i2 + i3)`  
= 2 / (5 + -3) = 2 / 2 = **1.0**  
→ Chia lấy số thực.

(d) `i1 // (i2 + i3)`  
= 2 // (5 + -3) = 2 // 2 = **1**  
→ Chia lấy phần nguyên.

(e) `i1 / i2 + i3`  
= 2 / 5 + (-3) = 0.4 + (-3) = **-2.6**  
→ Chia trước, cộng sau.

(f) `i1 // i2 + i3`  
= 2 // 5 + (-3) = 0 + (-3) = **-3**  
→ Chia lấy nguyên trước, cộng sau.

(g) `3 + 4 + 5 / 3`  
= 3 + 4 + (5 / 3) = 3 + 4 + 1.666... = **8.666...**  
→ Chia trước, cộng sau.

(h) `3 + 4 + 5 // 3`  
= 3 + 4 + (5 // 3) = 3 + 4 + 1 = **8**  
→ Chia lấy nguyên trước, cộng sau.

(i) `(3 + 4 + 5) / 3`  
= (12) / 3 = **4.0**  
→ Cộng trong ngoặc trước, chia sau.

(j) `(3 + 4 + 5) // 3`  
= (12) // 3 = **4**  
→ Cộng trong ngoặc trước, chia lấy nguyên.

(k) `d1 + (d2 * d3)`  
= 2.0 + (5.0 * -0.5) = 2.0 + (-2.5) = **-0.5**  
→ Nhân trước, cộng sau.

(l) `d1 + d2 * d3`  
= 2.0 + 5.0 * -0.5 = 2.0 + (-2.5) = **-0.5**  
→ Nhân trước, cộng sau.

(m) `d1 / d2 - d3`  
= 2.0 / 5.0 - (-0.5) = 0.4 + 0.5 = **0.9**  
→ Chia trước, trừ sau (trừ số âm thành cộng).

(n) `d1 / (d2 - d3)`  
= 2.0 / (5.0 - (-0.5)) = 2.0 / 5.5 = **0.3636...**  
→ Trừ trong ngoặc trước, chia sau.

(o) `d1 + d2 + d3 / 3`  
= 2.0 + 5.0 + (-0.5 / 3) = 2.0 + 5.0 + (-0.1666...) = **6.833...**  
→ Chia trước, cộng sau.

(p) `(d1 + d2 + d3) / 3`  
= (2.0 + 5.0 + -0.5) / 3 = (6.5) / 3 = **2.166...**  
→ Cộng trong ngoặc trước, chia sau.

(q) `d1 + d2 + (d3 / 3)`  
= 2.0 + 5.0 + (-0.5 / 3) = 2.0 + 5.0 + (-0.1666...) = **6.833...**  
→ Chia trong ngoặc trước, cộng sau.

(r) `3 * (d1 + d2) * (d1 - d3)`  
= 3 * (2.0 + 5.0) * (2.0 - (-0.5)) = 3 * 7.0 * 2.5 = 21.0 * 2.5 = **52.5**  
→ Cộng trong ngoặc trước, nhân sau.
